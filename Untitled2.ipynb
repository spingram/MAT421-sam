Write a function my_num_calc(f,a,b,n,option)
, where the output I
 is the numerical integral of f
, a function object, computed on a grid of n
 evenly spaced points starting at a
 and ending at b
. The integration method used should be one of the following strings defined by option: ‘rect’, ‘trap’, ‘left’,'right'. You may assume that n
 is odd.

```python
def my_num_calc(f,a,b,n,option):
      space = (b-a)/(n+1)
      points = [a + i * space for i in range(0,n+2)]
      area=0
      if option == 'right':
        for i in range(1,len(points)):
          area = area + f(points[i])*space
      elif option == 'rect':
        for i in range(0,len(points)-1):
          area = area + (f((points[i]+points[i+1])*.5))*space
      elif option == 'trap':
          for i in range(0,len(points)-1):
              area = area + (f(points[i])+f(points[i+1]))*space*.5
      else:
         for i in range(0,len(points)-1):
            area = area + f(points[i])*space

      return area


```

```python
f = lambda x: x**2

print("true value: 21.33")
left_est = my_num_calc(f,0,4,10,'left')
print("left estimation: " + str(left_est))
right_est = my_num_calc(f,0,4,10,'right')
print("right estimation: " + str(right_est))
mid_est = my_num_calc(f,0,4,10,'rect')
print("midpoint estimation: " + str(mid_est))
trap_est = my_num_calc(f,0,4,10,'trap')
print("trapazoid estimation: " + str(trap_est))

```

For x^2 show that the left and right sum converge to the true value as more squares are used.

```python
import matplotlib.pyplot as plt


n = [ i for i in range(1,100)]
left_est = [ 0 for i in range(1,100)]
right_est = [ 0 for i in range(1,100)]
true_val =  [ 21.33 for i in range(1,100)]
f = lambda x: x**2

for i in range(1,100):
  left_est[i-1] = my_num_calc(f,0,4,i,'left')
  right_est[i-1] = my_num_calc(f,0,4,i,'right')

plt.plot(n,left_est,label="Left estimation")
plt.plot(n,right_est,label = "Right estimation")
plt.plot(n,true_val, label = "True value")
plt.xlabel('Number of points')
plt.ylabel('Area')
plt.title("Change in Estimation Accuracy With Smaller Rectangles")
plt.legend()
plt.show()




```
